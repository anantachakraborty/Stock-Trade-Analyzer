```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the trade log data from the CSV file
trade_data = pd.read_csv('E:/Stock Trade Analyzer/data/trade_log.csv')
```

```python
# Histogram of Entry Price

# Create a figure with two subplots (1 row, 2 columns)
plt.figure(figsize=(12, 5))

# Subplot 1: Histogram of Entry Price
plt.subplot(1, 2, 1)
plt.hist(trade_data['Entry Price'], bins=20, color='skyblue')
plt.title('Histogram of Entry Price')
plt.xlabel('Entry Price')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Histogram of Exit Price

# Create a figure with two subplots (1 row, 2 columns)
plt.figure(figsize=(12, 5))

# Subplot 2: Histogram of Exit Price
plt.subplot(1, 2, 1)
plt.hist(trade_data['Exit Price'], bins=20, color='lightcoral')
plt.title('Histogram of Exit Price')
plt.xlabel('Exit Price')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Scatter Plot of Entry vs. Exit Price

# Create a figure
plt.figure(figsize=(8, 6))

# Create a scatter plot of Entry vs. Exit Price
plt.scatter(trade_data['Entry Price'], trade_data['Exit Price'], color='teal')

plt.title('Entry Price vs. Exit Price')
plt.xlabel('Entry Price')
plt.ylabel('Exit Price')
plt.grid()
plt.show()
```
