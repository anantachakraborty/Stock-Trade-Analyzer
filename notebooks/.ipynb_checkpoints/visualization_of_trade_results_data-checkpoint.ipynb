```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the trade results data from the CSV file
trade_data = pd.read_csv('E:/Stock Trade Analyzer/results/trade_results.csv')
```

```python
# Calculate and extract summary metrics
summary_metrics = [
    'Total Trades',
    'Profitable Trades',
    'Loss-Making Trades',
    'Win Rate',
    'Average Profit per Trade',
    'Average Loss per Trade',
    'Risk Reward Ratio',
    'Expectancy',
    'Average ROR per Trade',
    'Sharpe Ratio',
    'Max Drawdown',
    'Max Drawdown Percentage',
    'CAGR',
    'Calmar Ratio'
]

values = [trade_data[metric].values[0] for metric in summary_metrics]

# Create a horizontal bar plot with switched axes
plt.figure(figsize=(20, 14))
plt.barh(summary_metrics, values, color=['b', 'g', 'r', 'm', 'y', 'c', 'purple', 'darkcyan', 'darkorange', 'salmon', 'darkseagreen', 'dodgerblue', 'indigo', 'gold'])
plt.title('Summary Metrics')
plt.xlabel('Values')
plt.tight_layout()

# Show the bar plot
plt.show()
```

```python
# Calculate and extract summary metrics
total_trades = trade_data['Total Trades'].values[0]
profitable_trades = trade_data['Profitable Trades'].values[0]
loss_making_trades = trade_data['Loss-Making Trades'].values[0]
win_rate = trade_data['Win Rate'].values[0]
average_profit_per_trade = trade_data['Average Profit per Trade'].values[0]
average_loss_per_trade = trade_data['Average Loss per Trade'].values[0]

# Create a bar plot for summary metrics
labels = ['Total Trades', 'Profitable Trades', 'Loss-Making Trades', 'Win Rate', 'Average Profit per Trade', 'Average Loss per Trade']
values = [total_trades, profitable_trades, loss_making_trades, win_rate, average_profit_per_trade, average_loss_per_trade]

plt.figure(figsize=(10, 6))
plt.bar(labels, values, color=['b', 'g', 'r', 'm', 'y', 'c'])
plt.title('Summary Metrics')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the bar plot
plt.show()

```
